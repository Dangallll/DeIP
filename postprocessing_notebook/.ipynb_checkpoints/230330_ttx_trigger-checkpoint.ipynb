```python
import numpy as np
import matplotlib.pyplot as plt
import tifffile as tiff
import os
plt.style.use('default')
%matplotlib inline
```

```python
Trace1 = np.genfromtxt("TTX_after.csv", dtype = 'float', delimiter = ',')
Trace2 = np.genfromtxt("TTX_before.csv", dtype = 'float', delimiter = ',')
```

```python
Trace1  = Trace1[1:,1:].flatten()
Trace2 = Trace2[1:,1:].flatten()
```

```python
plt.figure(figsize = (12,8))
plt.plot(Trace1)
plt.plot(Trace2);
```

```python
def getDFF(pixel, Bin = 500, crop = (5001,79000), Q = 0.48):
    
    baseline = np.zeros(crop[1]-crop[0])
    
    for i in range(crop[0], crop[1]):
        
        window = pixel[i-Bin:i+Bin]
        base = np.quantile(window, Q)
        baseline[i-crop[0]] = base
    
    F0 = np.median(baseline)
    DFF = (pixel[crop[0]:crop[1]] - baseline) / F0
    return DFF.flatten()
```

```python
Trace1_flat = getDFF(Trace1)
Trace2_flat = getDFF(Trace2)
```

```python
plt.figure(figsize = (12,8))
plt.plot(Trace1_flat)
plt.plot(Trace2_flat)
```

```python
def window(curve, k, R1, R2):
    window = curve[k-R1:k+R2]
    return window


def STA1(spikelist, trace, plot, Normalize): 
    # spike frame list, int Channel, True/False, True/False
    
    R1 = 150
    R2 = 250

    peaks = np.zeros((R1+R2, len(spikelist)))
    
    for idx in range(len(spikelist)):
        
        peaks[:, idx] = window(trace, spikelist[idx], R1, R2)

    sta = np.mean(peaks, axis = 1)
    
    if Normalize == True:
        
        sta = sta / np.max(sta)
    
    if plot == True:
        plt.plot(sta)
    
    return peaks

def Findpeaks(curve, th, d_w = 600):
    spikelist = []
    for i in range(d_w,len(curve)-d_w):
        if all(curve[i] >= curve[i-d_w:i+d_w]) == True:
            if curve[i] > th:
                spikelist.append(i)
    return spikelist
```

```python
spikelist = Findpeaks(Trace2_flat, 0.02)
len(spikelist)
```

```python
plt.figure(figsize = (12,8))
trace = STA1(spikelist, Trace2_flat, True, False);
artifact = STA1(spikelist, Trace1_flat, True, False);
plt.xlabel("frame")
plt.ylabel("Df/f")
plt.show()

np.savetxt("peaks_beforTTX.csv", trace, delimiter = ',')
np.savetxt("peaks_afterTTX.csv", artifact, delimiter = ',')
```

```python
plt.figure(figsize = (12,8))
plt.plot(trace-artifact)
plt.xlabel("frame")
plt.ylabel("Df/f")
plt.show()
```
